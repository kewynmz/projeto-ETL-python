```python
# ==================================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# ==================================================

import pandas as pd
import seaborn as srn
import statistics  as sts
```

```python
# ==================================================
# CARREGAMENTO DOS DADOS
# ==================================================
dataset = pd.read_csv("dataset_funcionarios.csv", sep=",")
#visulizar
dataset.head()
```

```python
#tamanho
dataset.shape
```

ANÁLISE E TRATAMENTO DOS DADOS

```python
# ===================================
# ANÁLISE E TRATAMENTO - NOME
# ===================================
agrupado = dataset.groupby(['Nome']).size()
agrupado
```

```python
# removendo espaços 
dataset["Nome"] = dataset["Nome"].str.strip()
agrupado = dataset.groupby("Nome").size()
print(agrupado)
```

```python
# ===================================
# ANÁLISE E TRATAMENTO - IDADE
# ===================================
print(dataset["Idade"].dtype)
```

```python
# Valores únicos
print(dataset["Idade"].unique())
```

```python
# verificando valores nulos
print(dataset["Idade"].isnull().sum())
```

```python
# convertendo para número
dataset["Idade"] = pd.to_numeric(dataset["Idade"], errors="coerce")
```

```python
# estatísticas
print(dataset["Idade"].describe())

print("Média:", sts.mean(dataset["Idade"].dropna()))
print("Mediana:", sts.median(dataset["Idade"].dropna()))
```

```python
# removendo idades impossíveis
dataset = dataset[(dataset["Idade"] >= 18) & (dataset["Idade"] <= 70)]
```

```python
# substituindo nulos
mediana = sts.median(dataset["Idade"].dropna())
dataset["Idade"] = dataset["Idade"].fillna(mediana)
```

```python
# conferindo

print(dataset["Idade"].unique())
```

```python
# ===================================
# ANÁLISE E TRATAMENTO - Salário
# ===================================
print(dataset["Salario"].head(40))

print(dataset["Salario"].dtype)
```

```python
# se houver "R$", espaços e ponto de milhar:

dataset["Salario"] = (
    dataset["Salario"]
    .astype(str)
    .str.replace("R$", "", regex=False)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .str.strip()
)
```

```python
# convertendo para número
dataset["Salario"] = pd.to_numeric(dataset["Salario"], errors="coerce")
```

```python
# analisando
print(dataset["Salario"].describe())

print("Média:", sts.mean(dataset["Salario"].dropna()))
print("Mediana:", sts.median(dataset["Salario"].dropna()))
```

```python
# caso haja salário negativo
dataset = dataset[dataset["Salario"] >= 0]
```

```python
# preenchendo valores nulos
media = sts.mean(dataset["Salario"].dropna())

dataset["Salario"] = dataset["Salario"].fillna(media)
```

```python
# verificando
print(dataset["Salario"].unique())
```

```python
print(dataset["Salario"].describe())
```

```python
print(dataset["Salario"].isnull().sum())
```

```python
# ===================================
# ANÁLISE E TRATAMENTO - Estado
# ===================================
print(dataset["Estado"].unique())

print(dataset["Estado"].value_counts())
```

```python
# removendo espaços
dataset["Estado"] = dataset["Estado"].str.strip()
```

```python
# padronizando estados
dataset["Estado"] = dataset["Estado"].replace({
    "SP": "São Paulo",
    "São Paulo": "São Paulo",
    "SAO PAULO": "São Paulo",
    "SÃO PAULO": "São Paulo",
    "sao paulo": "São Paulo",

    "RIO DE JANEIRO": "Rio de Janeiro",
    "Rio de Janeiro": "Rio de Janeiro",

    "MG": "Minas Gerais",
    "Minas Gerais": "Minas Gerais",
    "MINAS GERAIS": "Minas Gerais",

    "PR": "Paraná",
    "Paraná": "Paraná",

    "SC": "Santa Catarina",
    "Santa Catarina": "Santa Catarina",

    "RS": "Rio Grande do Sul",
    "Rio Grande do Sul": "Rio Grande do Sul",

    "BA": "Bahia",
    "Bahia": "Bahia"
})
```

```python
# verificando
print(dataset["Estado"].unique())

print(dataset["Estado"].value_counts())
```

```python
# verificando nulos
dataset['Estado'].isnull().sum()
```

```python
# substituindo pela moda
dataset['Estado'].fillna('São Paulo', inplace=True)
```

```python
# verificando
print(dataset["Estado"].unique())

print(dataset["Estado"].value_counts())
```

```python
# ===================================
# ANÁLISE E TRATAMENTO - Data de admisão
# ===================================
print(dataset["Data_Admissao"].head(20))
```

```python
# convertendo as datas para o mesmo formato
dataset["Data_Admissao"] = pd.to_datetime(
    dataset["Data_Admissao"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)
```

```python
# conferindo
print(dataset["Data_Admissao"].isna().sum())
```

```python
# padronizando para dia/mês/ano
dataset["Data_Admissao"] = dataset["Data_Admissao"].dt.strftime("%d/%m/%Y")
```

```python
print(dataset["Data_Admissao"].head(20))
```

```python
# ===================================
# ANÁLISE E TRATAMENTO - Cargo
# ===================================
print(dataset["Cargo"].unique())

print(dataset["Cargo"].value_counts())
```

```python
# removendo espaços
dataset["Cargo"] = dataset["Cargo"].str.strip()
```

```python
# verificando nulos
print(dataset["Cargo"].isna().sum())
```

```python
dataset.to_csv("dataset_funcionarios_tratado_final.csv", index=False)
```
